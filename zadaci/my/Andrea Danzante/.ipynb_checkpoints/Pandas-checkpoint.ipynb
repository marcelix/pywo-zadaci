**6) U stupac odst upisati odstupanje od prosjeka**

```python
df.head()
```

```python
df["odst"] = 0.0
df.head()
```

```python
prosjeci = df.groupby("mon").aggregate(np.mean).avg.values
```

```python
def f_odst(m,a):
    return a - prosjeci(m-1)
```

```python
df.odst = f_odst(df.avg, df.mon)
```

```python
for i in range(1,13):
    df.odst = df[df.mon == i].avg - prosjeci[i-1]
```
