<h2><em style="font-style:normal;text-decoration:underline">Zadatak 1.1</em> - Euklidov algoritam</h2>

<p>a) Generator koji vraća ostatke dijeljenja (prva 2 člana su sami brojevi koji se ispituju)

```python
def Euklidov_algoritam (prvi,drugi):
    if drugi>prvi:
        prvi,drugi=drugi,prvi
    yield prvi
    yield drugi
    while prvi%drugi!=0:
        prvi,drugi=drugi,(prvi%drugi)
        yield drugi
list(Euklidov_algoritam(112,13))
```

<p>b) Generator koji vraća kvocijente tokom izvršavanja Euklidovog algoritma</p>

```python
def Euklidov_algoritam (prvi,drugi):
    if drugi>prvi:
        prvi,drugi=drugi,prvi
    yield prvi//drugi
    while prvi%drugi!=0:
        prvi,drugi=drugi,(prvi%drugi)
        yield prvi//drugi
list(Euklidov_algoritam(112,13))  #u primjeru je greška (zadnji broj mora biti 2, a ne 1)
```

<p>c) Funkcija koja ovisno o proslijeđenim parametrima ispisuje najveće zajedničke mjere unesenih brojeva (može ih biti više od 2) ili ostatke pri dijeljenju pri traženju NZM-a svih brojeva</p>

```python
def Euklidov_algoritam (*brojevi,**kwargs):
    if kwargs["nzm"]!="True" and kwargs["nzm"]!="False":
        print("Unos mora biti tipa: 56 112 7,nzm=False|True")
        exit()
    nzm=brojevi[0]
    for i in range (1,len(brojevi)):
        broj=brojevi[i]
        if nzm>brojevi[i]:
            nzm,broj=broj,nzm #u varijablu "nzm" će se pohraniti manji broj od 2 uspoređivanih
        while broj%nzm!=0:
            broj,nzm=nzm,broj%nzm
            if kwargs["nzm"]=="False" and nzm!=0:
                yield nzm
        if kwargs["nzm"]=="True":
            yield nzm

def fja_koja_poziva_inputbox_za_generatore_temeljene_nad_euklidovim_algoritmom():
    str_argumenti=list(argument for argument in str(input("Unesite niz brojeva razdvojenih zarezima (npr. 56 112 7,nzm=False|True): ")).split(","))
    if len(str_argumenti)!=2:
        print("Unos mora biti tipa: 56 112 7,nzm=False|True")
        return
    str_argumenti[0]=str_argumenti[0].strip() #u slučaju da korisnik unese razmak ispred/iza zareza
    str_argumenti[1]=str_argumenti[1].strip()

    str_argumenti[0]=[int(x) for x in str_argumenti[0].split(" ")]
    imenik={ str(str_argumenti[1].split("=")[0]) : str(str_argumenti[1].split("=")[1]) }

    if str_argumenti[1]=="nzm=True":
        print("Najveće zajedničke mjere zadanih brojeva (počevši od prva 2 člana, pa tada između njihove mjere i trećeg,...): \r\n" +
              ", ".join([str(i) for i in Euklidov_algoritam (*(str_argumenti[0]),**imenik)]))
    elif str_argumenti[1]=="nzm=False":
        print("Brojevi koji su ostaci modularnog dijeljenja tijekom izvršavanja Euklidovog algoritma i različiti od 0: " + "\r\n" +
             ", ".join([str(i) for i in Euklidov_algoritam (*(str_argumenti[0]),**imenik)]))
    else:
        print("Unos mora biti tipa: 56 112 7,nzm=False|True")
```

<h2><em style="font-style:normal;text-decoration:underline">Zadatak 1.1</em> - Hornerov algoritam</h2>

<p>a) Generator koji vraća elemente retka Hornerove sheme</p>

```python
def Hornerov_algoritam(nultocka,koeficijenti):
    yield koeficijenti[0]
    for i in range (1,len(koeficijenti)): #išlo bi od 0, ali prvi član ionako prepisuje vrijednost vodećeg člana
        koeficijenti[i]=nultocka*koeficijenti[i-1] + koeficijenti[i]
        yield koeficijenti[i]
[x for x in Hornerov_algoritam(2,[1,2,-4,7])]
```

<p>b) Generator koji vraća retke Hornerove sheme

```python
def Hornerov_algoritam(nultocka,koeficijenti):
    while len(koeficijenti)>0:
        for i in range (1,len(koeficijenti)): #išlo bi od 0, ali prvi član ionako prepisuje vrijednost vodećeg člana
            koeficijenti[i]=nultocka*koeficijenti[i-1] + koeficijenti[i]
        yield list(koeficijenti)
        del koeficijenti[-1]

[x for x in Hornerov_algoritam(2,[1,2,-4,7])]
```

<p>c) Funkcija koja vraća traži cjelobrojne kandidate za nultočke polinoma, ispituje svakog od njih pomoću Hornerovog algoritma, te se u konačnici pomoću dobivenih cjelobrojnih nultočaka (ako ih uopće ima) ispiše polinom nakon faktorizacije (faktorizirani polinom se ispiše i u LaTeX formatu)</p>

```python
polinom=str(input("Unesite koeficijente polinoma razdvojene razmacima: "))
koeficijenti=polinom.split(" ")
potencijalna_rjesenja=[]

koeficijenti=list(int(x) for x in koeficijenti)

for i in range (1,int(abs(koeficijenti[-1])/2)+1):
    if koeficijenti[-1] % i==0:
        potencijalna_rjesenja.append(i)
        potencijalna_rjesenja.append(-i)
        potencijalna_rjesenja.append(koeficijenti[-1]//i) #rezultat je ionako isti kao da se obično dijeli (sa /),
        potencijalna_rjesenja.append(-koeficijenti[-1]//i) #no onda vraća cjelobrojne brojeve u float formi (3.0
                                                            #umjesto 3, što čini konačan ispis neprivlačnijim)
        
def Hornerov_algoritam(koeficijenti,nultocka):
    pomocna=koeficijenti[:]
    kratnost=0
    while True:
        for i in range (1,len(koeficijenti)): #išlo bi od 0, ali prvi član ionako prepisuje vrijednost vodećeg člana
            pomocna[i]=nultocka*pomocna[i-1] + pomocna[i]
        if pomocna[-1]==0:
            del pomocna[-1]
            kratnost+=1
            koeficijenti=pomocna[:]
        else:
            return (koeficijenti,kratnost)
#            break
#    return kratnost
nultocke=[]
faktorizirani_polinom=""
for clan in potencijalna_rjesenja:
    ntorka=Hornerov_algoritam(koeficijenti,clan)#slicingom se funkciji prosljeđuje nova lista (tj. formalni
        #argument poprima nezavisni skup brojeva, dok bi se bez nje proslijedila referenca na listu koja je
        #proslijeđena čime bi se pri svakom daljnom pozivu proslijeđivala sasvim druga lista)
    kratnost=ntorka[1]
    if kratnost>0:
        koeficijenti=ntorka[0][:]
        if -clan>0:
            faktorizirani_polinom+="(x+"+str(-clan)+")"
        elif -clan<0:
            faktorizirani_polinom+="(x"+str(-clan)+")"
        else: #-clan==0
            faktorizirani_polinom+="x"
        if kratnost>1:
            faktorizirani_polinom+="^"+str(kratnost)+" "
        for i in range (0,kratnost):
            nultocke.append(str(clan))
if len(koeficijenti)>1:
    faktorizirani_polinom+="("
    for i in range(0,len(koeficijenti)):
        potencija=len(koeficijenti)-i-1
        if koeficijenti[i]>0:
            if i>0: #kod koeficijenta prvog člana nije potrebno prikazati predstojeći pozitivni predznak
                faktorizirani_polinom+="+"
            if koeficijenti[i]!=1:
                faktorizirani_polinom+=str(koeficijenti[i])
            if potencija>0:
                faktorizirani_polinom+="x"
                if potencija>1:
                    faktorizirani_polinom+="^"+str(potencija)
        elif koeficijenti[i]<0:
            faktorizirani_polinom+=str(koeficijenti[i])
            if koeficijenti[i]==-1:
                del faktorizirani_polinom[-1] #isti efekt kao i u prethodnom if dijelu
            if potencija>0:
                faktorizirani_polinom+="x"
                if potencija>1:
                    faktorizirani_polinom+="^"+str(potencija)
        #inače (ako je koeficijenti[i]==0), tada izostavi taj član
    faktorizirani_polinom+=")"
print("Cjelobrojne nultočke zadanog polinoma: " + ", ".join(nultocke) \
+ "\r\nFaktorizirani polinom sa cjelobrojnim nultočkama: "+faktorizirani_polinom)
```

<h2><em style="font-style:normal;text-decoration:underline">Zadatak 1.2</em> - Generator koji ispisuje <i>po redu</i> (prema Cantorovoj zamisli) racionalne brojeve čiji zbroj brojnika i nazivnika je jednak ili manji od $n$</h2>

```python
from IPython.display import display, Math, Latex

def najveca_zajednicka_mjera(a,b):
    if (a<b):
        a,b=b,a
    while (a%b!=0):
        a,b=b,a%b
    return b

def generator_racionalnih_brojeva(zbroj_clanova_razlomka=10): #zbroj brojnika i nazivnika
    if zbroj_clanova_razlomka>0:
        yield 0
    i=2
    while (True):
        if zbroj_clanova_razlomka>=i:
            for j in range(1,i):
                if j%(i-j)!=0:
                    nzm=najveca_zajednicka_mjera(j,i-j)
    #                if nzm!=1: #ako se broj da pokratiti, onda je taj broj već izgeneriran (jer po tome suma članova
                                #razlomka daje broj manji od trenutnog brojača "i")
    #                    if j/nzm + (i-j)/nzm < i:
    #                       break
    #                    else
    #                        yield [j/nzm,(i-j)/nzm]
    #                        yield [-j/nzm,(i-j)/nzm]
                    if nzm==1:
                        yield [j,i-j]
                        yield [-j,i-j]
                else:
                    if (i-j == 1): #ako se razlomak može skratiti da bude cijeli broj (npr. 8/4==2), tada
                                    #ga nema smisla ponovno generirati jer je već prethodno dodan
                                    # 8/4 -> 8+4=12=i (11. iteracija) 2=2/1 -> 2+1=3=i (u 2. iteraciji je bio dodan)
                        yield j//(i-j)  #koristi se cjelobrojno dijeljenje jer su ti brojevi ionako djeljivi, pa da
                        yield -j//(i-j) #ostatak bude tipa integer, a ne float
            i+=1
        else:
            break

string=""
for num in generator_racionalnih_brojeva(10):
    if type(num)==int:
#        display(Math(str(num)))
        string+=str(num)+", "
    else: # if type(num)==list:
        string+='\\frac{'+str(num[0])+'}{'+str(num[1])+'}, '
string=string[:-2]
display(Math(string))
```

<h2><em style="font-style:normal;text-decoration:underline">Zadatak 1.2</em> - Usavršeni program koji vraća sve permutacije određenog znakovnog niza (bilo koje veličine!)</h2>

```python
permutacije=[]
def funkcija(pozicija,niz):
    for i in range (len(niz)-1,pozicija-1,-1):
        if pozicija!=-1:
            lista=niz[:]
            lista[i],lista[pozicija]=lista[pozicija],lista[i]
            funkcija(pozicija-1,lista[:])
        else:
            permutacije.append(niz)
            return
        
funkcija(len(list("matematika"))-1,list("matematika"))
len(permutacije)  #broj permutacija sa ponavljanjem (jer se u riječi 'matematika' nalaze slova koja se ponavljaju)
```

```python
skup=set()
for perm in permutacije:
    skup.add("".join(perm))
len(skup)  #broj permutacija bez ponavljanja
```

```python
", ".join(sorted(list(skup)))  #sortirani popis svih permutacija (bez ponavljanja)
```

<p>b) Sortiranje (prez)imena sudionika prema hrvatskoj abecedi</p>

```python
letters="abcčćdđefghijklmnopqrsštuvwxyzž"
rjecnik={slovo:letters.index(slovo) for slovo in letters}

def fja_usporedbe(string1,string2,i=0):
    if rjecnik[string1[i]]<rjecnik[string2[i]]:
        return -1
    elif rjecnik[string1[i]]>rjecnik[string2[i]]:
        return 1
    else:
        if i==len(string1)-1 and i==len(string2)-1:
            return 0
        elif i==len(string1)-1 and i!=len(string2)-1:
            return 1
        elif i!=len(string1)-1 and i==len(string2)-1:
            return -1
        else:
            return fja_usporedbe(string1,string2,i+1)

def nadfja(lista1,lista2):
    rezultat=fja_usporedbe(lista1[0].lower(),lista2[0].lower())
    if rezultat!=0:
        return rezultat
    else:
        return fja_usporedbe(lista1[1].lower(),lista2[1].lower())
    
def sortiranje_izborom(lista):
    for i in range(0,len(lista)-1):
        for j in range (len(lista)-1,i,-1):
            if nadfja(lista[i],lista[j])>0:
                lista[i],lista[j]=lista[j],lista[i]
    return lista

sortiranje_izborom(L[:])  #tako "lista" i dalje sadrži nepotpuno sortirani skup sudionika
```

<p>b) Sortiranje sudionika prema hrvatskoj abecedi<p>
<span style="text-decoration:underline;font-weight:bold">POBOLJŠANJA:</span>
<ul>
    <li>Hrvatski digrami se također tretiraju u skladu sa hrvatskim jezikoslovnim pravilima</li>
    <li>U slučaju da su prezimena osoba jednaka, daljnja usporedba se vrši prema njihovim imenima</li>
    <li>Velika i mala slova se tretiraju na jednak način: Vlado $\ngtr$ ana</li>
</ul>

```python
letters="abcčćdđefghijklmnopqrsštuvwxyzž"
rjecnik={slovo:letters.index(slovo) for slovo in letters}

def vrati_odgovarajucu_vrijednost_ovisno_o_znaku(string,i):
    if i!=len(string)-1:
        if string[i]=="l" and string[i+1]=="j":
            return rjecnik["l"]+0.5
        elif string[i]=="n" and string[i+1]=="j":
            return rjecnik["n"]+0.5
        elif string[i]=="d" and string[i+1]=="ž":
            return rjecnik["d"]+0.5
        else:
            return rjecnik[string[i]]
    else:
        return rjecnik[string[i]]

def fja_usporedbe(string1,string2,i=0):
    vrijednost1=vrati_odgovarajucu_vrijednost_ovisno_o_znaku(string1,i)
    vrijednost2=vrati_odgovarajucu_vrijednost_ovisno_o_znaku(string2,i)
    
    if vrijednost1<vrijednost2:
        return -1
    elif vrijednost1>vrijednost2:
        return 1
    else:
        if i==len(string1)-1 and i==len(string2)-1:
            return 0
        elif i==len(string1)-1 and i!=len(string2)-1:
            return 1
        elif i!=len(string1)-1 and i==len(string2)-1:
            return -1
        else:
            return fja_usporedbe(string1,string2,i+1)

def nadfja(lista1,lista2):
    rezultat=fja_usporedbe(lista1[0].lower(),lista2[0].lower())
    if rezultat!=0:
        return rezultat
    else:
        return fja_usporedbe(lista1[1].lower(),lista2[1].lower())
    
def sortiranje_izborom(lista):
    for i in range(0,len(lista)-1):
        for j in range (len(lista)-1,i,-1):
            if nadfja(lista[i],lista[j])>0:
                lista[i],lista[j]=lista[j],lista[i]
    return lista

L.append(["Njegovan","Karlo"])
L.append(["Novak","Petar"])
L.append(["Ljerković","Jerko"])
L.append(["Lucin","Jasna"])
L.append(["Šestak","Ivan"])
L.append(["Šestak","Juraj"])
sortiranje_izborom(L[:])
```

<h2><em style="font-style:normal;text-decoration:underline">Zadatak 2.2</em></h2>

<p>a) Dovršite zadatak s radionice tako da je klase podržavaju ovakav račun:
$$a+b, a-b, 2*b, 4c$$
pri čemu su $a, b, c$ vektori.</p>
<p>b) Snimanje obiju klasa u datoteku <em>ravnina.py</em></p>

```python
# %%writefile ravnina.py

import math #from math import sqrt  - time se uključuje samo ono što nam treba (čak je i bolja od ove obične)

class Tocka: #class naziv_klase(naslijedjena_klasa1,naslijedjena_klasa2,...)
    """
        Opis ove klase.
        2-dim tocke u ravnini.
    """
    def __init__(self,x,y): #ovo je konstruktor klase - u zagradama se mogu navesti argumenti koji se proslijeđuju
                    #objektu pri njegovom alociranju
        self.x=x
        self.y=y
    def norm(self):
        return math.sqrt(self.x**2+self.y**2)
    def __repr__(self):
        return "Tocka({0}, {1})".format(self.x,self.y)
    
class Vektor(Tocka):   #klasa vektor naslijeđuje atribute i metode od klase Tocka
    def __add__(self,b):
        """
        self je vektor, b je vektor
        """
        return Vektor(self.x+b.x , self.y+b.y)
    def __sub__(self,b):
        return Vektor(self.x-b.x , self.y-b.y)
    def __mul__(self,b):
        return Vektor(self.x*b , self.y*b)
    def __rmul__(self,b): #metoda za realizaciju množenja vektora skalarom kad se skalar nalazi ispred vektora
        #return Vektor(self.x*b , self.y*b)#time je moguće izračunati 2*a, a ne samo a*2 kao što je sa __mul__ moguće
        return Vektor.__mul__(self,b)
    def __repr__(self):
        return "Vektor({0}, {1})".format(self.x,self.y)
```

```python
a = Vektor(1,2)
b = Vektor(-12,34)
c = Vektor(34,-13)
```
